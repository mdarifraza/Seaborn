## strip plot
# ðŸ“Œ Seaborn Strip Plot â€” Interview Notes

---

## ðŸŸ¦ 1. What is a Strip Plot?
A **strip plot** displays **individual data points** of a categorical variable along a single axis.  
It is similar to a scatterplot but used for **categorical vs numerical** data.

Used for:
- Showing distribution of raw data points  
- Identifying spread and outliers  
- Comparing categories  

Often combined with **boxplot** or **violin plot** for deeper understanding.

---

## ðŸŸ© 2. Syntax
```python
sns.stripplot(x='category', y='value', data=df)
```
---

# ðŸŸ¦ 11. Interview Points
- Strip plot shows **raw data distribution** for categorical variables  
- Helps to view **data spread, clusters, and outliers**  
- Uses jitter to reduce overlapping  
- Works well with **small and medium datasets**  
- Often combined with boxplot or violin plot  
- Good for comparing multiple categories  

---



````python
# axes level function
sns.stripplot(data=tips,x='day',y='total_bill')
````

````python
# figure level function
sns.catplot(kind='strip',data=tips,x='day',y='total_bill')
````

````python
#jitter
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',jitter=0.2,hue='sex')
````

### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (â€œminimumâ€, first quartile [Q1], median, third quartile [Q3] and â€œmaximumâ€). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

# ðŸ“Œ Seaborn Boxplot â€” Interview Notes

---

## ðŸŸ¦ 1. What is a Box Plot?
A **boxplot** shows the **distribution of numerical data** using five-number summary:
- Minimum  
- Q1 (25th percentile)  
- Median (50th percentile)  
- Q3 (75th percentile)  
- Maximum  

It also highlights **outliers** clearly.

Used for:
- Comparing distributions  
- Detecting outliers  
- Understanding spread & skewness  

---

## ðŸŸ© 2. Syntax
```python

```

---

# ðŸŸ¦ 11. Interview Points
- Boxplot shows **spread, skewness, and outliers**  
- Based on **5-number summary**  
- Outliers appear as individual points  
- Good for comparing multiple categories  
- Works best for medium/large datasets  
- Can combine with stripplot/swarmplot for more clarity  
- Helps detect:
  - asymmetry  
  - extreme values  
  - interquartile range (IQR)  

---



````python
#box plots
sns.boxplot(data=tips,x='day',y='total_bill')
````

## violinplot=(boxplot+KDE plot)
# ðŸ“Œ **Seaborn Violinplot â€” Interview Notes**

---

## ðŸŸ¦ 1. **What is a Violin Plot?**

A **violinplot** combines:

- **Boxplot** â†’ shows median, quartiles, IQR  
- **KDE Plot** â†’ shows smooth probability density  

It visualizes:

- Data distribution  
- Spread & density  
- Skewness  
- Multiple peaks (multi-modal distribution)  

Used for:

- Comparing distributions  
- Understanding shape & density  
- Better visualization than boxplot alone  

---

## ðŸŸ© 2. **Syntax**

```python
sns.violinplot(x='category', y='value', data=df)
```

---

# ðŸŸ¦ 11. **Interview Points**

- Violin plot = **Boxplot + KDE plot**  
- Shows **distribution shape**, not just summary  
- Captures **multi-modal** distributions  
- Split violin helps compare sub-groups  
- Good for skewed / non-normal data  
- More informative than boxplot for density  
- Shows **IQR, median, spread, density** all in one  
- Works well with large datasets  

---


````python
# violinpplot
sns.violinplot(data=tips,x='day',y='total_bill')
````

````python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
````

````python
#hue
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex')
````

## barplots
# ðŸ“Œ **Seaborn Barplot â€” Interview Notes**

---

## ðŸŸ¦ 1. **What is a Barplot?**

A **barplot** in Seaborn shows the **average (mean)** value of a numerical variable for each category.

It automatically:

- Calculates **mean** of the numeric column  
- Shows **confidence intervals (CI)**  
- Displays comparisons between categories  

Used for:

- Category-wise comparison  
- Showing average values  
- Highlighting trends or differences  

---

## ðŸŸ© 2. **Syntax**

```python
sns.barplot(x='category', y='value', data=df)
```
---

# ðŸŸ¦ 11. **Interview Points**

- Barplot shows **mean values** by default  
- Supports **custom estimators** (median, count, sum)  
- Confidence interval (CI) included by default  
- Good for comparing **groups or categories**  
- Works best when you want to show **summary statistics**, not raw values  
- Different from `countplot`:  
  - **barplot â†’ shows mean**  
  - **countplot â†’ shows frequency counts**  

---


````python
sns.barplot(data=tips,x='sex',y='total_bill')
````

````python
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker')
````

````python
#estimator
import numpy as np
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator=np.min)
````

````python
#remove the bar
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator=np.min,errorbar=None)
````

## point plot
# ðŸ“Œ **Seaborn Pointplot â€” Interview Notes**

---

## ðŸŸ¦ 1. **What is a Point Plot?**

A **pointplot** shows **aggregated values of a numerical variable** for each category as **points connected by lines**.  

It is similar to **barplot**, but better for showing **trends over categories**.

Used for:

- Visualizing **mean or other estimator values**  
- Showing **trends between categories**  
- Comparing groups with **confidence intervals**  

---

## ðŸŸ© 2. **Syntax**

```python
sns.pointplot(x='category', y='value', data=df)
```
---

# ðŸŸ¦ 11. **Interview Points**

- Pointplot shows **trends using points + lines**  
- Aggregates values using **mean by default**  
- Supports **custom estimators** (median, sum, etc.)  
- Confidence intervals (CI) show variability  
- Ideal for **trend comparison between categories**  
- Similar to barplot but **line connects the points** for easier trend visualization  

---


````python
sns.pointplot(data=tips,x='sex',y='total_bill',hue='smoker')
````

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars

## count plot
# ðŸ“Œ **Seaborn Countplot â€” Interview Notes**

---

## ðŸŸ¦ 1. **What is a Count Plot?**

A **countplot** shows the **frequency (count) of observations** in each categorical variable.  

It is useful for:

- Visualizing **category distribution**  
- Detecting **imbalanced classes**  
- Quick **summary of categorical data**  

Unlike **barplot**, it **does not require a numeric variable**, it counts occurrences automatically.

---

## ðŸŸ© 2. **Syntax**

```python
sns.countplot(x='category', data=df)
```

---

# ðŸŸ¦ 9. **Interview Points**

- Countplot **counts occurrences** of categories  
- Automatically computes frequency â€” no numeric column needed  
- Supports **hue, palette, order, orientation**  
- Useful for **EDA** and **imbalanced datasets**  
- Different from barplot:  
  - **barplot â†’ shows summary statistic of numeric variable**  
  - **countplot â†’ shows frequency counts**  
- Works well with **categorical features**  

---


````python
sns.countplot(data=tips,x='sex',hue='day')
````

A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable

### A second way to plot Facet plots -> FacetGrid
# ðŸ“Œ **Seaborn FacetGrid â€” Interview Notes**

---

## ðŸŸ¦ 1. **What is a FacetGrid?**

A **FacetGrid** in Seaborn allows you to **visualize data across multiple subplots** (facets) based on **row, column, or hue categories**.

It is useful for:

- Comparing **distributions or relationships** across subsets of data  
- Splitting data into **rows, columns, or colors**  
- Creating **small multiples** for visual analysis  

---

## ðŸŸ© 2. **Syntax**

```python
g = sns.FacetGrid(data=df, row='row_cat', col='col_cat', hue='hue_cat')
g.map(sns.scatterplot, 'x', 'y')
```
---

# ðŸŸ¦ 9. **Interview Points**

- FacetGrid is for **multi-plot visualizations**  
- Supports **row, column, hue** to create facets  
- Useful for **small multiples** in EDA  
- `.map` â†’ applies a plotting function to each facet  
- `.map_dataframe` â†’ passes the entire DataFrame to plotting function  
- Helps **compare patterns** across subsets of data  
- Works with histplot, scatterplot, boxplot, kdeplot, etc.  

---


````python
# figure level -> relplot -> displot -> catplot -> lmplot
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
````

````python
## facetgrid
g=sns.FacetGrid(data=tips,col='day',row='time',hue='smoker')
g.map(sns.violinplot,'sex','total_bill')
````

### Plotting Pairwise Relationship (PairGrid Vs Pairplot)
# ðŸ“Œ **Seaborn PairGrid vs Pairplot â€” Interview Notes**

---

## ðŸŸ¦ 1. **What is PairGrid?**

A **PairGrid** is a **grid for plotting pairwise relationships** in a dataset.  

- Gives **full control** over plots for **each combination of variables**  
- You can map **different plots** to **upper, lower, and diagonal** parts of the grid  
- Used for **custom pairwise visualizations**  

---

## ðŸŸ© 2. **Syntax (PairGrid)**

```python
g = sns.PairGrid(df)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
```

- `map_upper` â†’ upper triangle plots  
- `map_lower` â†’ lower triangle plots  
- `map_diag` â†’ diagonal plots  

---

## ðŸŸ© 4. **What is Pairplot?**

A **Pairplot** is a **simplified wrapper around PairGrid**.  

- Automatically plots **scatterplots** for pairwise relationships  
- Diagonal shows **histograms or KDE plots**  
- Less flexible than PairGrid but faster to create  

---

## ðŸŸ¦ 5. **Syntax (Pairplot)**

```python
sns.pairplot(df, hue='sex', kind='scatter', diag_kind='hist')
plt.show()
```

- `hue` â†’ color by category  
- `kind` â†’ scatter / reg (regression)  
- `diag_kind` â†’ hist / kde  

---

## ðŸŸ© 6. **Key Differences (PairGrid vs Pairplot)**

| Feature            | PairGrid                       | Pairplot                       |
|-------------------|--------------------------------|--------------------------------|
| Flexibility        | High (custom plots for upper, lower, diag) | Low (predefined scatter/hist) |
| Use Case           | Detailed/custom pairwise analysis | Quick pairwise visualization  |
| Diagonal Plots     | Can choose individually        | Auto (hist or kde)            |
| Syntax Complexity  | More verbose                   | One-liner                     |
| Hue Support        | Needs `map_*` with hue manually | Built-in                      |

---

## ðŸŸ¦ 7. **Interview Points**

- **PairGrid** â†’ full control, more complex, flexible  
- **Pairplot** â†’ quick, easy, less control  
- Both are useful for **pairwise relationships** and **EDA**  
- Pairplot = **PairGrid wrapper** with default settings  

---


````python
# pairplot
sns.pairplot(iris)
````

````python
sns.pairplot(iris,hue='species')
````

````python
# pair grid
g=sns.PairGrid(data=iris,hue='species')
g.map(sns.scatterplot)
````

````python
g=sns.PairGrid(data=iris,hue='species')
g.map(sns.scatterplot)
````

````python
# map_diag -> map_offdiag
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
````

````python
# map_diag -> map_upper -> map_lower
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
````

````python
# vars
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
````

### JointGrid Vs Jointplot
# ðŸ“Œ **Seaborn JointGrid vs Jointplot â€” Interview Notes**

---

## ðŸŸ¦ 1. **What is JointGrid?**

A **JointGrid** is a **flexible grid for plotting a bivariate relationship** between two variables along with **their marginal distributions**.  

- Allows **full control** over **main plot** and **marginal plots**  
- You can map **different plot types** for main and marginal axes  
- Useful for **custom bivariate analysis**

---

## ðŸŸ© 2. **Syntax (JointGrid)**

```python
g = sns.JointGrid(x='total_bill', y='tip', data=tips)
g.plot(sns.scatterplot, sns.histplot)
```

- `g.plot(main_func, marginal_func)`  
  - `main_func` â†’ main bivariate plot  
  - `marginal_func` â†’ marginal distributions  

---
---

## ðŸŸ© 4. **What is Jointplot?**

A **Jointplot** is a **simplified wrapper around JointGrid**.  

- Automatically creates a **bivariate plot with marginals**  
- Less flexible but **faster to implement**  
- Supports scatter, regression, hex, kde plots  

---

## ðŸŸ¦ 5. **Syntax (Jointplot)**

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
plt.show()
```

- `kind` options: `scatter`, `reg`, `kde`, `hex`  

---

## ðŸŸ© 6. **Key Differences (JointGrid vs Jointplot)**

| Feature           | JointGrid                        | Jointplot                         |
|------------------|----------------------------------|----------------------------------|
| Flexibility       | High (custom main + marginals)   | Low (predefined plots)           |
| Use Case          | Detailed/custom bivariate analysis | Quick bivariate visualization    |
| Plot Customization| Map any plot function individually | Limited to `kind` options        |
| Syntax Complexity | More verbose                     | One-liner                        |
| Marginals         | Fully controlled                 | Auto generated                   |

---

## ðŸŸ¦ 7. **Interview Points**

- **JointGrid** â†’ full control, customizable  
- **Jointplot** â†’ quick and easy wrapper around JointGrid  
- Both show **bivariate relationships + marginal distributions**  
- Use JointGrid for **advanced/custom visualizations**  
- Use Jointplot for **quick EDA / simple plots**  

---


````python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hist',hue='sex')
````

````python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='kde',hue='sex')
````

````python
#join grid
g=sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.kdeplot,sns.violinplot)
````
