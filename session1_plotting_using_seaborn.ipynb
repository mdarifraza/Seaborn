## scatterplot
# ðŸ“Œ Seaborn Scatterplot â€” Interview Notes

---

## ðŸŸ¦ 1. What is a Scatterplot?
A **scatterplot** displays the relationship between two numerical variables  
by plotting points (x, y) on a graph.

Used for:
- Pattern detection  
- Correlation  
- Outliers  
- Trend analysis  

---

## ðŸŸ© 2. Syntax
```python
sns.scatterplot(x='col1', y='col2', data=df)
```

---
## ðŸŸ¦ 11. Interview Points
- Used for relationship between **two numeric variables**
- Seaborn scatterplot = **statistically enhanced Matplotlib scatterplot**
- Supports:
  - hue (color)
  - size (bubble effect)
  - style (different markers)
  - palettes
  - legend control
- Works great with Pandas DataFrames  
- `relplot` is a wrapper that can create scatter + facet plots

---



````python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px     # only use inbuilt dataset
````

````python
tips=sns.load_dataset('tips')
tips
````

````python
# scatter plot ->axes level function
sns.scatterplot(data=tips,x='total_bill',y='tip')
````

````python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex')
````

````python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time')
````

````python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
````

````python
#relplot->figure level function
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter')
````

````python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
````

## histogram
# ðŸ“Œ Seaborn Histogram â€” Interview Notes

---

## ðŸŸ¦ 1. What is a Histogram?
A **histogram** shows the **frequency distribution** of a numerical variable by dividing data into bins.

Used for:
- Understanding distribution  
- Detecting skewness  
- Outliers  
- Spread of data  

---

## ðŸŸ© 2. Syntax
```python
sns.histplot(data)
```

---


---

## ðŸŸ© 12. Interview Points
- Histogram shows **distribution of numeric data**
- `histplot` = replacement for deprecated `distplot`
- Supports:
  - bins  
  - kde  
  - hue  
  - multiple (overlay, stack, fill)
- Works on arrays + DataFrame columns  
- `displot` is figure-level: supports faceting

---




````python
# plotting univariate histogram
sns.histplot(data=tips,x='total_bill')
````

````python
sns.displot(data=tips,x='total_bill',kind='hist')
````

````python
# bins 
sns.displot(data=tips,x='total_bill',kind='hist',bins=2)
````

````python
# Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram. 
# Discrete bins are automatically set for categorical variables

# countplot
sns.displot(data=tips,x='day',kind='hist')
````

````python
#hue parameter
sns.displot(data=tips,x='tip',kind='hist',hue='sex')
````

````python
#element
sns.displot(data=tips,x='tip',kind='hist',hue='sex',element='step')
````

## kde plot
# ðŸ“Œ Seaborn KDE Plot â€” Interview Notes

---

## ðŸŸ¦ 1. What is a KDE Plot?
A **KDE (Kernel Density Estimation) plot** shows a **smooth continuous probability density curve**  
based on the distribution of a numerical variable.

It is basically a **smoothed version of a histogram**.

Used for:
- Understanding data distribution  
- Detecting peaks  
- Comparing multiple distributions  
- Smooth density visualization  

---

## ðŸŸ© 2. Syntax
```python
sns.kdeplot(data)
```
---

## ðŸŸ© 12. Interview Points
- KDE = Smoothed version of histogram  
- Shows **probability density**, not frequency  
- Best for continuous variables  
- `kdeplot` supports:
  - hue  
  - fill  
  - bivariate density  
  - bw_adjust  
- Avoid KDE for **small datasets** or **data with ma**


````python
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(data=tips,x='total_bill')
````

````python
sns.displot(data=tips,x='total_bill',kind='kde')
````

````python
# hue->fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex')
````

````python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
````

## rugplot
# ðŸ“Œ Seaborn Rugplot â€” Interview Notes

---

## ðŸŸ¦ 1. What is a Rug Plot?
A **rugplot** displays **small vertical/horizontal lines** (ticks) along an axis,  
each representing one data point.

Used for:
- Showing **distribution of raw data points**
- Adding context to histogram/KDE plots  
- Showing **data density** visually  

It is often combined with KDE or histogram.

---

## ðŸŸ© 2. Syntax
```python
sns.rugplot(data)
```

---

# ðŸŸ¦ 11. Interview Points
- Rugplot shows **actual data points** as small lines  
- Good for small datasets  
- Usually combined with **KDE** and **histograms**  
- Shows **density concentration** visually  
- Works on arrays + Pandas columns  
- Simple but powerful for understanding raw distribution  

---



````python
# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
````

````python
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
````

````python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot 
# and then shows the count of observations within each rectangle with the fill color

# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data=tips,x='total_bill',y='tip',kind='hist')
````

````python
#Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.displot(data=tips, x='total_bill', y='tip',kind='kde')
````

### 2. Matrix Plot
A matrix plot is a graphical representation of data arranged in a 2-dimensional grid (rows Ã— columns), where each cell is displayed using colors, symbols, or numbers to show the value at that position.

- It is mainly used to visualize:

## heatmap
# ðŸ“Œ Seaborn Heatmap â€” Interview Notes

---

## ðŸŸ¦ 1. What is a Heatmap?
A **heatmap** shows data in a **matrix form**, where values are represented using **color intensity**.

Used for:
- Correlation analysis  
- Matrix visualization  
- Pattern detection  
- Missing value visualization  

---

## ðŸŸ© 2. Syntax
```python
sns.heatmap(data)
```
---

# ðŸŸ© 12. Interview Points
- Heatmap is best for **matrix-style data**  
- Commonly used for **correlation analysis**  
- `annot=True` shows numeric values  
- `cmap` controls color scheme  
- `linewidths` adds separation lines  
- Works great for:
  - correlation matrices  
  - pivot tables  
  - missing value visualization  
- Helps identify patterns & relationships quickly  

---



````python
# Plot rectangular data as a color-encoded matrix
temp_df=gap.pivot(index='country',columns='year',values='lifeExp')
temp_df
````

````python
# axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
````

````python
# annot
temp_df=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
temp_df
````

````python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5,cmap='summer')
````

## clustermap
# ðŸ“Œ Seaborn Clustermap â€” Interview Notes

---

## ðŸŸ¦ 1. What is a Clustermap?
A **clustermap** is a **heatmap + hierarchical clustering** combined.

It:
- Groups similar rows and columns together  
- Uses **dendrograms** (tree-like structures) to show clusters  
- Helps identify patterns and similarities in data  

Used for:
- Feature clustering  
- Correlation clustering  
- Pattern recognition  
- High-dimensional data visualization  

---

## ðŸŸ© 2. Syntax
```python
sns.clustermap(data)
```

---

# ðŸŸ¦ 13. Interview Points
- Clustermap = **Heatmap + hierarchical clustering**
- Automatically reorders rows & columns based on similarity  
- Displays **dendrograms** to show cluster hierarchy  
- `standard_scale` and `z_score` used for normalization  
- Great for:
  - similarity detection  
  - correlation clustering  
  - feature relationships  
- More powerful than a simple heatmap  

---



````python
# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.
````
